```python
from scipy.misc.pilutil import imread, imsave, imshow
import matplotlib.pyplot as plt
import numpy as np

# read in image as 8 bit grayscale
img1= imread('Penguins.jpg', mode ='L')
img2= imread('Koala.jpg', mode ='L')

# specify a threshold 0-255
threshold = 50

# make all pixels < threshold black
binarized1 = 1.0 * (img1 > threshold)
binarized2 = 1.0 * (img2 > threshold)
# save the binarized image
imsave('binarna_slika.jpg', binarized1)
imsave('binarna_koala.jpg', binarized2)
#imshow(binarized)

plt.imshow(binarized1)
plt.show()


```

```python
from __future__ import print_function
import os
from PIL import Image

files = [
      'Penguins.jpg',
      'Koala.jpg',
      'binarna_slika.jpg',
     'binarna_koala.jpg']

result = Image.new("RGB", (1200, 1200))

for index, file in enumerate(files):
    path = os.path.expanduser(file)
    img = Image.open(path)
    img.thumbnail((600, 800), Image.ANTIALIAS)
    x = index // 2 * 600
    y = index % 2 * 600
    w, h = img.size
    result.paste(img, (x, y, x + w, y + h))

result.save(os.path.expanduser('output.jpg'))
plt.imshow(result)
plt.show()
```
